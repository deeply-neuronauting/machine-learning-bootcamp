```python
def impute_age(cols):
    Age = cols['Age']
    Pclass = cols['Pclass']
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train["Age"] = train[["Age", "Pclass"]].apply(impute_age, axis=1)
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

#### **drop_first=True** to solve multicolinearty
That parameter tells pandas:

Drop the first category to avoid the dummy variable trap.

In the Titanic dataset, "Sex" has 2 categories:

female

male

When you run:

python
Copy code
pd.get_dummies(train["Sex"], drop_first=True)
Pandas does this:

Keeps only the second category alphabetically  "male"

Drops "female"

So internally:

Sex	male (dummy)
female	0
male	1

In below version:

female  False

male  True



```python
sex = pd.get_dummies(train["Sex"], drop_first=True).astype(int)
sex
```

```python
embark = pd.get_dummies(train["Embarked"], drop_first=True)
embark
```

```python
train = pd.concat([train, sex, embark], axis=1)
train.head(5)
```

```python
train.drop(["Sex", "Embarked", "Name", "Ticket"], axis=1, inplace=True, errors='ignore')
train.head(5)
```
